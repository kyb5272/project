# **Open Source SW and Python Programming Project: Implementation of Subway Navigation**

In this project, you will implement a navigation system for Seoul subway stations, especially **step by step**




# 0. Initialization

Read subway station information by reading the file **simplified_subway_info.xlsx**
*  The excel file contains subway station information of Seoul subway line 1 ~ 4
*  When you execute this sourcecode, you MUST upload these files in your Colab runtime environment
*   **MUST NOT** change this Initialization code cell

```
import xlrd

# Read data file
data = xlrd.open_workbook("simplified_subway_info.xlsx")
data = data.sheet_by_name('Sheet1')

# Store the loaded book object as a string list in subwayStation variable
subwayStation = []
for line in range(4) :
    cur = [x for x in data.col_values(line) if x]    
    subwayStation.append(cur[1:])
```

# 1. Your implementation
Let's start the implementation of a subway naviation system by using the loaded subway station information :)

*   You can access the subway station information by referring to **subwayStation** variable (list type)

```
keys = ['line1', 'line2', 'line3', 'line4']             # 호선과 역 이름 dictionary로 변환
stations = subwayStation
subwayStation = {}
for line, station in zip(keys, stations):
    subwayStation[line] = tuple(station)

separator = '*' * 50
errorMsg = '>>>> [ERROR] {}'



def dispSubwaylineInfo() :                              # option1 
    print(separator)
    print("Subway line information service")
    print(separator)
    while True :
        line = int(input(">> Please enter a subway line number (1 - 4): "))
        if line < 1 or line > 4:
            print(errorMsg.format('Please enter a valid subway line number.'))
            print('\n')
        else :
            line = 'line' + str(line)
            print(">>", subwayStation[line])
            print('\n')
            break




def dispSubwayStationInfo() :                           # option 2
    print(separator)
    print("Subway station information service")
    print(separator)
    while True :
        station = input(">> please enter a subway station name: ")
        check = 0                                                                                     

        for b in range(4):
            stationNames = subwayStation['line'+str(b+1)]
            for c in stationNames :
                if c==station :
                    print(station, "station is in line "+str(b+1))
                    check =1
        if check == 1 :
            print("\n")
            break
        elif check == 0 :
            print(errorMsg.format('Please enter a valid station name'))
            print("\n")




def findPath() :                                        # option 3
    linenum1=set()      # 출발역의 호선 set형식
    linenum2=set()      # 도착역의 호선 set형식
    
    while True :        # 출발역 호선확인
        DepartureStation = input("Please enter your departure station :")
        check = 0
        for b in range(4):
            stationNames = subwayStation['line'+str(b+1)]
            for c in stationNames :
                if c==DepartureStation :                # 출발역 출력
                    print(DepartureStation, "DepartureStation is in line "+str(b+1))
                    linenum1.add(b+1)                   # 출발역의 호선 linenum1에 저장
                    check = 1
        if check == 1 :
            print("\n")
            break
        elif check == 0 :
            print(errorMsg.format('Please enter a valid DepartureStation'))
            print('\n')

    while True :        # 도착역 호선확인
        ArrivalStation = input("Please enter your arrival station :")
        check = 0
        for b in range(4):
            stationNames = subwayStation['line'+str(b+1)]
            for c in stationNames :
                if c==ArrivalStation :                  # 도착역 출력
                    print(ArrivalStation, "ArrivalStation is in line "+str(b+1))
                    linenum2.add(b+1)                   # 도착역의 호선 linenum2에 저장
                    check = 1
        if check == 1 :
            print("\n")
            break
        elif check == 0 :
            print(errorMsg.format('Please enter a valid DepartureStation'))
            print('\n')

    
    if linenum1.isdisjoint(linenum2):   # 출발역과 도착역이 서로 다른 호선
        linenum1 = list(linenum1)       # set형식의 출발역 호선 list화
        linenum2 = list(linenum2)       # set형식의 도착역 호선 list화
        setStation1 = set(subwayStation['line'+str(linenum1[0])])       # 출발역호선의 모든 역 이름 set형식으로 저장
        setStation2 = set(subwayStation['line'+str(linenum2[0])])       # 도착역호선의 모든 역 이름 set형식으로 저장
        
        transferstationList = setStation1 & setStation2         # 환승역 찾기
        k = list(transferstationList)       # set형식의 환승역 list화
        transferstation = k[0]              # 여러 환승역 중 제일 처음의 환승역만 사용
        
        a = subwayStation['line'+str(linenum1[0])].index(DepartureStation)   # 출발역 호선의 출발역 index 저장
        b = subwayStation['line'+str(linenum1[0])].index(transferstation)    # 출발역 호선의 환승역 index 저장
        c = subwayStation['line'+str(linenum2[0])].index(transferstation)    # 도착역 호선의 환승역 index 저장
        d = subwayStation['line'+str(linenum2[0])].index(ArrivalStation)     # 도착역 호선의 도착역 index 저장
        
        if a < b:                   # 출발역과 환승역 사이의 모든 역 출력
            print(subwayStation['line'+str(linenum1[0])][a:b+1])
            if c < d :              # 환승역과 도착역 사이의 모든 역 출력
                print(subwayStation['line'+str(linenum2[0])][c:d+1])
            else :
                if d==0:
                    print(subwayStation['line'+str(linenum2[0])][c::-1]) 
                else :
                    print(subwayStation['line'+str(linenum2[0])][c:d-1:-1]) 
        else :
            if b==0 :
                print(subwayStation['line'+str(linenum1[0])][a::-1])
            else :
                print(subwayStation['line'+str(linenum1[0])][a:b-1:-1])
            if c < d :
                print(subwayStation['line'+str(linenum2[0])][c:d+1])
            else :
                if d==0:
                    print(subwayStation['line'+str(linenum2[0])][c::-1])
                else :
                    print(subwayStation['line'+str(linenum2[0])][c:d-1:-1])  

    else :               # 출발역과 도착역이 서로 같은 호선 
        set(linenum1)
        set(linenum2)
        sameline = linenum1 & linenum2          # 출발역과 도착역이 무슨 호선으로 같은지 sameline에 저장
        sameline = list(sameline)
        for b in range(4):
            if b+1 == sameline[0] :  # 호선 찾기 
                line = 'line' + str(b+1)
                a = subwayStation[line].index(DepartureStation)         # 같은 호선에서 출발역의 index값을 a에 저장
                b = subwayStation[line].index(ArrivalStation)           # 같은 호선에서 도착역의 index값을 b에 저장
                if a < b:                                       #출발역과 도착역 사이의 모든 역 출력
                    print(subwayStation[line][a:b+1])
                else :
                    if b==0:
                        print(subwayStation[line][a::-1])
                    else :
                        print(subwayStation[line][a:b-1:-1])       






while True :
    print(separator)
    print("1. Display subway line information (Line 1 ~ 4)")
    print("2. Display subway station information")
    print("3. Find a path between two subway stations")
    print("4. Exit")
    print(separator)
    option = int(input("Please choose one of the options (1 ~ 4): "))
    print('\n')
    if option==1 :
        dispSubwaylineInfo()
    
    elif option==2 :
        dispSubwayStationInfo()
    
    elif option==3 :
        findPath()
    
    elif option==4 :
        print("The end")
        break

    else:
        print(errorMsg.format('Please choose a valid option.'))
```
